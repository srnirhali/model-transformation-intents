```python
import re
import tensorflow as tf

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

padding_type='pre'
oov_tok = "<OOV>"
vocab_size = 10000
embedding_dim = 16
max_length = 100
```

```python
# remove symbols and line breaks

# remove line breaks
training_sentences = [re.sub(' +', ' ', text.replace('\n',' ')) for text in training_sentences]
validation_sentences = [re.sub(' +', ' ', text.replace('\n',' ')) for text in validation_sentences]
testing_sentences = [re.sub(' +', ' ', text.replace('\n',' ')) for text in testing_sentences]

# remove symbols
training_sentences = [re.sub(' +', ' ', re.sub(r'[^\w\s]', ' ', text)) for text in training_sentences]
validation_sentences = [re.sub(' +', ' ', re.sub(r'[^\w\s]', ' ', text)) for text in validation_sentences]
testing_sentences = [re.sub(' +', ' ', re.sub(r'[^\w\s]', ' ', text)) for text in testing_sentences]

# remove numbers
training_sentences = [re.sub(' +', ' ', re.sub(r'[^a-zA-Z]', ' ', text)) for text in training_sentences]
validation_sentences = [re.sub(' +', ' ', re.sub(r'[^a-zA-Z]', ' ', text)) for text in validation_sentences]
testing_sentences = [re.sub(' +', ' ', re.sub(r'[^a-zA-Z]', ' ', text)) for text in testing_sentences]

print(training_sentences[0])
```

```python
from nltk.corpus import stopwords  
from nltk.tokenize import word_tokenize

stop_words = set(stopwords.words('english'))  

print(stop_words)

stop_words.remove('from')
stop_words.remove('to')
stop_words.remove('if')
stop_words.remove('then')
stop_words.remove('in')
stop_words.remove('out')

print(stop_words)

  
word_tokens = [sentence.split() for sentence in training_sentences]
training_sentences = [' '.join([w for w in word_sent if not w in stop_words]) for word_sent in word_tokens]

word_tokens = [sentence.split() for sentence in validation_sentences]
validation_sentences = [' '.join([w for w in word_sent if not w in stop_words]) for word_sent in word_tokens]

word_tokens = [sentence.split() for sentence in validation_sentences]
testing_sentences = [' '.join([w for w in word_sent if not w in stop_words]) for word_sent in word_tokens]

print(training_sentences[0])
```

```python
# Divide the parameters in the text
training_sentences = [re.sub(' +', ' ', ' '.join([t.lower() if t.isupper() else (' '.join(re.findall('[A-Z][^A-Z]*', t))).lower() for t in list(text.split())])) for text in training_sentences]
validation_sentences = [re.sub(' +', ' ', ' '.join([t.lower() if t.isupper() else (' '.join(re.findall('[A-Z][^A-Z]*', t))).lower() for t in list(text.split())])) for text in validation_sentences]
testing_sentences = [re.sub(' +', ' ', ' '.join([t.lower() if t.isupper() else (' '.join(re.findall('[A-Z][^A-Z]*', t))).lower() for t in list(text.split())])) for text in testing_sentences]

print(training_sentences[0])
```

```python
from nltk.stem.porter import PorterStemmer
porter = PorterStemmer()

training_sentences = [' '.join([porter.stem(word) for word in text.split()]) for text in training_sentences]
validation_sentences = [' '.join([porter.stem(word) for word in text.split()]) for text in validation_sentences]
testing_sentences = [' '.join([porter.stem(word) for word in text.split()]) for text in testing_sentences]

print(training_sentences[0])
```

```python
# Remove duplicate words
training_sentences = [' '.join(sorted(list(set(text.split())))) for text in training_sentences]
validation_sentences = [' '.join(sorted(list(set(text.split())))) for text in validation_sentences]
testing_sentences = [' '.join(sorted(list(set(text.split())))) for text in testing_sentences]

print(training_sentences[0])
```

```python
tokenizer = Tokenizer(num_words=vocab_size, oov_token=oov_tok)
tokenizer.fit_on_texts(training_sentences)

word_index = tokenizer.word_index

training_sequences = tokenizer.texts_to_sequences(training_sentences)
training_padded = pad_sequences(training_sequences, maxlen=max_length, padding=padding_type)

validation_sequences = tokenizer.texts_to_sequences(validation_sentences)
validation_padded = pad_sequences(validation_sequences, maxlen=max_length, padding=padding_type)

testing_sequences = tokenizer.texts_to_sequences(testing_sentences)
testing_padded = pad_sequences(testing_sequences, maxlen=max_length, padding=padding_type)

print(training_sequences[0])

print(word_index)
```

```python
import numpy as np
training_padded = np.array(training_padded)
training_labels = np.array(training_labels)
validation_padded = np.array(validation_padded)
validation_labels = np.array(validation_labels)
testing_padded = np.array(testing_padded)
testing_labels = np.array(testing_labels)

print(training_padded[0])
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(vocab_size, embedding_dim, trainable=True, input_length=max_length),
    # tf.keras.layers.Conv1D(32, 5, activation='relu'),
    # tf.keras.layers.MaxPooling1D(5),
    # tf.keras.layers.Conv1D(64, 5, activation='relu'),
    # tf.keras.layers.MaxPooling1D(10),
    # tf.keras.layers.Flatten(),
    tf.keras.layers.GlobalAveragePooling1D(),
    # tf.keras.layers.Dense(1024, activation='relu'),
    # tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    # tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(9, activation='softmax')
])
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
num_epochs = 300
from tensorflow.keras.utils import to_categorical
train_labels = to_categorical(training_labels, 9)
val_labels = to_categorical(validation_labels, 9)
history = model.fit(training_padded, train_labels, epochs=num_epochs, validation_data=(validation_padded, val_labels), batch_size=2)
```

```python
import matplotlib.pyplot as plt


def plot_graphs(history, string):
  plt.plot(history.history[string])
  plt.plot(history.history['val_'+string])
  plt.xlabel("Epochs")
  plt.ylabel(string)
  plt.legend([string, 'val_'+string])
  plt.show()
  
plot_graphs(history, "accuracy")
plot_graphs(history, "loss")
```
